```python
# Importing required libraries
import pandas as pd, datetime
```

```python
#Read in the data files 
test_file = "resources/test.csv"
test_df = pd.read_csv(test_file)
grocery_file="resources/grocery_train.csv"
grocery_train_df=pd.read_csv(grocery_file)
grocery_train_df.head(5)
holiday_file="resources/holidays_events.csv"
holidays_events_df = pd.read_csv(holiday_file)
oil_file= "resources/oil.csv"
oil_df= pd.read_csv(oil_file)
```

```python
# merge grocery_train_df and holidays_events_df on date
merged_df = pd.merge(grocery_train_df , holidays_events_df, on='date', how='left').merge(oil_df, on='date', how='left')
merged_df
```

```python
# create day, week, month, and year columns
merged_df['month'] = pd.to_datetime(merged_df['date']).dt.month
merged_df['year'] = pd.to_datetime(merged_df['date']).dt.year
merged_df['day'] = pd.to_datetime(merged_df['date']).dt.day
merged_df['week'] = pd.to_datetime(merged_df['date']).dt.week
merged_df.head()
```

```python
# saving the dataframe
merged_df.to_csv('new_training.csv')
```
